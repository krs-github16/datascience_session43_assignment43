## Task 2 
Given a list of strings - List[String] (alpha, gamma, omega, zeta, beta) - 
    
    find count of all strings with length 4 - 
    convert the list of string to a list of integers, where each string is mapped to its corresponding length - 
    find count of all strings which contain alphabet m - 
    find the count of all strings which start with the alphabet a

```python
strings=['alpha','agamma','omega','zeta','beta']
print(strings)
```

```python
count_4=0
for string in strings:
    if len(string)==4:
        count_4+=1
print(f'count of all strings with length 4 is {count_4}') #count of all strings with length 4
        
```

```python
len_s=[]
for string in strings:
    len_s.append(len(string))
    
print(f'string is mapped to its corresponding length and the list is {len_s}') #each string is mapped to its corresponding length
```

```python
count_a=0
for string in strings:
    if string[0].lower()=='a':
        count_a+=1
print(f'count of all strings which start with the alphabet a is {count_a}') #count of all strings which start with the alphabet a
        
```

```python
count_m=0
for string in strings:
    for s in string:
        if s.lower()=='m':
            count_m+=1
            break;
print(f'count of all strings which contain alphabet m is {count_m}') #count of all strings which contain alphabet m
```
